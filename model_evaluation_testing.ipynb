```python
import glob
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2

model = YOLO(glob.glob('/kaggle/input/yolo_model/keras/default/1/best (1).pt')[0])


# Path to the image you want to test
image_path = '/kaggle/input/images/TEST_IMAGES/7.jpg'

# Perform prediction
out = model.predict(image_path, save=True, conf=0.2)

```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load and display image
img = mpimg.imread('/kaggle/input/images/TEST_IMAGES/7.jpg')
plt.imshow(img)
plt.axis('off')  # Hide axes
plt.show()
```
